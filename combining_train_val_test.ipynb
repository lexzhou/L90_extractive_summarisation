```python
import json
import pandas as pd 

data_train = json.load(open('./data/train.json'))
data_val = json.load(open('./data/validation.json'))
data_test = json.load(open('./data/test.json'))

summary = []
article = []
ID = []
data_type = []

for row in data_train:
    summary.append(row["summary"])
    article.append(row["article"])
    ID.append(row["id"])
    data_type.append("train")
    
for row in data_val:
    summary.append(row["summary"])
    article.append(row["article"])
    ID.append(row["id"])
    data_type.append("validation")
    
for row in data_test:
    summary.append(row["summary"])
    article.append(row["article"])
    ID.append(row["id"])
    data_type.append("test")
    
# Create a dictionary with your data
data = {
    'summary': summary,
    'article': article,
    'id': ID,
    'data_type': data_type
}

# Create the DataFrame
df = pd.DataFrame(data)
df.to_csv("./stanford-corenlp-full-2018-02-27/all_data.csv", index=False)
```
